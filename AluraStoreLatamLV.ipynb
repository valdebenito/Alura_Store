### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



```python
total_precio_tienda1 = tienda['Precio'].sum()
total_precio_tienda2 = tienda2['Precio'].sum()
total_precio_tienda3 = tienda3['Precio'].sum()
total_precio_tienda4 = tienda4['Precio'].sum()

total_precios = [
    ('Tienda 1', total_precio_tienda1),
    ('Tienda 2', total_precio_tienda2),
    ('Tienda 3', total_precio_tienda3),
    ('Tienda 4', total_precio_tienda4)
]

print("Total precio por tienda:")
for tienda_nombre, total in total_precios:
    print(f"{tienda_nombre}: {total:.2f}")
```

```python
import matplotlib.pyplot as plt

tiendas = [tienda[0] for tienda in total_precios]
totales = [tienda[1] for tienda in total_precios]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, totales, color=['blue', 'green', 'red', 'purple'])
plt.xlabel("Tienda")
plt.ylabel("Total de Venta")
plt.title("Total de Venta por Tienda")
plt.show()
```

```python
import matplotlib.pyplot as plt

tiendas = [tienda[0] for tienda in total_precios]
totales = [tienda[1] for tienda in total_precios]

plt.figure(figsize=(8, 8))
plt.pie(totales, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=['cyan', 'green', 'red', 'purple'])
plt.title("Distribución de Venta por Tienda")
plt.show()
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("Cantidad de productos vendidos por categoría en Tienda 1:")
display(ventas_por_categoria_tienda1)

print("\nCantidad de productos vendidos por categoría en Tienda 2:")
display(ventas_por_categoria_tienda2)

print("\nCantidad de productos vendidos por categoría en Tienda 3:")
display(ventas_por_categoria_tienda3)

print("\nCantidad de productos vendidos por categoría en Tienda 4:")
display(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda1 = tienda['Calificación'].mean()
promedio_calificacion_tienda2 = tienda2['Calificación'].mean()
promedio_calificacion_tienda3 = tienda3['Calificación'].mean()
promedio_calificacion_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {promedio_calificacion_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {promedio_calificacion_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {promedio_calificacion_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {promedio_calificacion_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
ventas_por_producto_tienda1 = tienda.groupby('Producto').size().sort_values(ascending=False)
ventas_por_producto_tienda2 = tienda2.groupby('Producto').size().sort_values(ascending=False)
ventas_por_producto_tienda3 = tienda3.groupby('Producto').size().sort_values(ascending=False)
ventas_por_producto_tienda4 = tienda4.groupby('Producto').size().sort_values(ascending=False)

print("Productos más vendidos en Tienda 1:")
display(ventas_por_producto_tienda1.head())

print("\nProductos menos vendidos en Tienda 1:")
display(ventas_por_producto_tienda1.tail())

print("\nProductos más vendidos en Tienda 2:")
display(ventas_por_producto_tienda2.head())

print("\nProductos menos vendidos en Tienda 2:")
display(ventas_por_producto_tienda2.tail())

print("\nProductos más vendidos en Tienda 3:")
display(ventas_por_producto_tienda3.head())

print("\nProductos menos vendidos en Tienda 3:")
display(ventas_por_producto_tienda3.tail())

print("\nProductos más vendidos en Tienda 4:")
display(ventas_por_producto_tienda4.head())

print("\nProductos menos vendidos en Tienda 4:")
display(ventas_por_producto_tienda4.tail())
```

```python
import matplotlib.pyplot as plt

# Visualize top 5 products per store
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Top 5 Productos Vendidos por Tienda', y=1.02)

ventas_por_producto_tienda1.head().plot(kind='bar', ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Tienda 1')
axes[0, 0].set_xlabel('Producto')
axes[0, 0].set_ylabel('Cantidad Vendida')

ventas_por_producto_tienda2.head().plot(kind='bar', ax=axes[0, 1], color='green')
axes[0, 1].set_title('Tienda 2')
axes[0, 1].set_xlabel('Producto')
axes[0, 1].set_ylabel('Cantidad Vendida')

ventas_por_producto_tienda3.head().plot(kind='bar', ax=axes[1, 0], color='red')
axes[1, 0].set_title('Tienda 3')
axes[1, 0].set_xlabel('Producto')
axes[1, 0].set_ylabel('Cantidad Vendida')

ventas_por_producto_tienda4.head().plot(kind='bar', ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Tienda 4')
axes[1, 1].set_xlabel('Producto')
axes[1, 1].set_ylabel('Cantidad Vendida')

plt.tight_layout()
plt.show()

# Visualize bottom 5 products per store
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Bottom 5 Productos Vendidos por Tienda', y=1.02)

ventas_por_producto_tienda1.tail().plot(kind='bar', ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Tienda 1')
axes[0, 0].set_xlabel('Producto')
axes[0, 0].set_ylabel('Cantidad Vendida')

ventas_por_producto_tienda2.tail().plot(kind='bar', ax=axes[0, 1], color='green')
axes[0, 1].set_title('Tienda 2')
axes[0, 1].set_xlabel('Producto')
axes[0, 1].set_ylabel('Cantidad Vendida')

ventas_por_producto_tienda3.tail().plot(kind='bar', ax=axes[1, 0], color='red')
axes[1, 0].set_title('Tienda 3')
axes[1, 0].set_xlabel('Producto')
axes[1, 0].set_ylabel('Cantidad Vendida')

ventas_por_producto_tienda4.tail().plot(kind='bar', ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Tienda 4')
axes[1, 1].set_xlabel('Producto')
axes[1, 1].set_ylabel('Cantidad Vendida')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

*   **Tienda 3:** Muebles y Electrónicos son las más vendidas, con Instrumentos musicales y Artículos para el hogar siendo las menos vendidas.
*   **Tienda 4:** Muebles y Electrónicos son las categorías principales, con Instrumentos musicales y Libros siendo las menos vendidas.

Todas las tiendas muestran patrones similares en las categorías más vendidas. Las diferencias en las categorías menos vendidas podrían indicar áreas de oportunidad o, por el contrario, justificar la venta si ciertas categorías de bajo rendimiento son un factor clave.

**4. Productos más y menos vendidos:**

Hemos identificado los productos individuales con mayor y menor volumen de ventas en cada tienda. Analizando estos productos específicos se podría entender mejor qué impulsa las ventas en cada ubicación y si los productos de bajo rendimiento en una tienda son consistentemente bajos en otras. Si una tienda tiene una alta concentración de productos de bajo rendimiento en comparación con las otras, podría ser un factor para su venta.

Sin embargo, si el enfoque es la venta, un costo de envío más alto en otras tiendas podría ser visto como un área a mejorar por el nuevo propietario, o simplemente un factor de costo que afecta la rentabilidad actual.

**Conclusión:**

Considerando todos los factores:

*   La **Tienda 4** tiene los **menores ingresos totales**.
*   Las **categorías de productos** más y menos vendidas son **similares** entre las tiendas, lo que no ofrece una clara distinción para la venta basada solo en esto.
*   Las **calificaciones promedio** son **parejas**, con la Tienda 3 ligeramente por encima.
*   El análisis detallado de los **productos individuales** más y menos vendidos es necesario para una conclusión más sólida en este punto.
*   La **Tienda 4** tiene el **costo de envío promedio más bajo**.

Basándonos principalmente en los ingresos totales, la **Tienda 4** podría ser el candidato más obvio para la venta si el objetivo es desprenderse de la tienda con el menor rendimiento financiero. Sin embargo, una decisión final debería considerar la estrategia general de la empresa y la importancia de cada uno de los factores analizados. Por ejemplo, si la rentabilidad (teniendo en cuenta los costos de envío) es un factor más importante que los ingresos brutos, la Tienda 3 o 2 con costos de envío más altos podrían ser consideradas. Además, una inmersión más profunda en la rentabilidad por categoría y por producto proporcionaría una imagen más completa.
